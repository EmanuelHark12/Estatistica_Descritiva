**28.a)** Descreva a distribuição da variável consumo de oxigênio (VO2) em cada um dos quatro momentos de avaliação, utilizando medidas resumo (minímo, máximo, quartis, mediana, média, desvio padrão, etc), boxplots e histogramas. Você identifica algum paciente com valores de consumo de oxigênio discrepantes? Interprete os resultados

```python
#Pegando as colunas de interesse

dataset = pd.read_excel('esforco.xls', sheet_name=2)
```

```python
#medidas resumo das variáveis

dataset.describe().round(1)
```

```python
#definindo estilo dos plots

sns.set_style('dark')
```

```python
plt.figure(figsize=(6,8))
sns.boxplot(y = 'Repouso', data= dataset)
plt.tick_params(labelsize = 12)
plt.ylabel('Consumo de Oxigênio (ml/(kg.min))', {'fontsize': 14})
plt.title('Consumo de oxigênio em momento de repouso', {'fontsize':16})
```

```python
plt.figure(figsize=(6,8))
sns.boxplot(y = 'LAN', data= dataset)
plt.tick_params(labelsize = 12)
plt.ylabel('Consumo de Oxigênio (ml/(kg.min))', {'fontsize': 14})
plt.title('Consumo de oxigênio em momento de limiar anaeróbio', {'fontsize':16})
```

```python
plt.figure(figsize=(6,8))
sns.boxplot(y = 'PCR', data= dataset)
plt.tick_params(labelsize = 12)
plt.ylabel('Consumo de Oxigênio (ml/(kg.min))', {'fontsize': 14})
plt.title('Consumo de oxigênio em ponto de compensação respiratória', {'fontsize':15})
```

```python
plt.figure(figsize=(6,8))
sns.boxplot(y = 'PICO', data= dataset)
plt.tick_params(labelsize = 12)
plt.ylabel('Consumo de Oxigênio (ml/(kg.min))', {'fontsize': 14})
plt.title('Consumo de oxigênio em momento de Pico', {'fontsize':16})

```

```python
plt.figure( figsize = (12,6))
plt.ylabel('Frequência', {'fontsize': 14})
plt.title('Consumo de oxigênio em momento de repouso', {'fontsize': 16})
sns.distplot(dataset['Repouso'], kde = False, color = 'mediumblue', bins = np.arange(1,7,0.4))
plt.xlabel('Consumo de oxigênio (ml/(kg.min))', {'fontsize': 14})
plt.xlim([1,7])
plt.xticks(np.arange(1, 8, 0.8))
plt.tick_params(labelsize = 12)
```

```python
plt.figure( figsize = (12,6))
plt.ylabel('Frequência', {'fontsize': 14})
plt.title('Consumo de oxigênio em momento de limiar anaeróbio', {'fontsize': 16})
sns.distplot(dataset['LAN'].dropna(), kde = False, color = 'mediumblue', bins = np.arange(5,23,1.2))
plt.xlabel('Consumo de oxigênio (ml/(kg.min))', {'fontsize': 14})
plt.xticks(np.arange(5,23,2.4))
plt.xlim([4,23])
plt.tick_params(labelsize = 12)
```

```python
plt.figure( figsize = (12,6))
plt.ylabel('Frequência', {'fontsize': 14})
plt.title('Consumo de oxigênio em ponto de compensação respiratória', {'fontsize': 16})
sns.distplot(dataset['PCR'].dropna(), kde = False, color = 'mediumblue', bins = np.arange(6,30,1.6))
plt.xlabel('Consumo de oxigênio (ml/(kg.min))', {'fontsize': 14})
plt.xticks(np.arange(6,31,3.2))
plt.xlim([5,31])
plt.tick_params(labelsize = 12)
```

```python
plt.figure( figsize = (12,6))
plt.ylabel('Frequência', {'fontsize': 14})
plt.title('Consumo de oxigênio em momento de pico', {'fontsize': 16})
sns.distplot(dataset['PICO'], kde = False, color = 'mediumblue', bins = np.arange(5,44,2.4))
plt.xlabel('Consumo de oxigênio (ml/(kg.min))', {'fontsize': 14})
plt.xticks(np.arange(5,44,4.8))
plt.xlim([4,43])
plt.tick_params(labelsize = 12)
```

**Resposta 28.a)** Podemos identificar algumas medidas discrepantes nos quatro gráficos boxplot. Além disso, notamos que a distribuição do consumo de oxigênio é semelhante nos quatro momentos em que foi medido, o consumo de oxigênio tende a subir de acordo com momento, sendo o Picoo momento com valores mais altos e Repouso o com valores mais baixos. 

**28.b)** Descreva a distribuição da classe funcional NYHA por meio de uma tabela de frequˆencias. Utilize um método gráfico para representar essa tabela.

```python
#Lendo os dados necessários

NYHA = pd.read_excel('esforco.xls', sheet_name=3)
```

```python
#removendo valores nulos

data = NYHA[NYHA['NYHA'] != 0]
```

```python
data['NYHA'].value_counts()
```

```python
#Gráfico de barras para o número de pacientes em cada classe funcional


plt.figure(figsize=(10,6))
ax = sns.countplot(x='NYHA', data=data)
plt.ylabel('Contagem', fontdict = {'fontsize':14})
plt.tick_params(labelsize = 12)
plt.title('Número de pacientes em cada classe funcional', fontdict = {'fontsize': 16})

for p in ax.patches:
    ax.annotate('{}'.format(p.get_height()), (p.get_x()+0.35, p.get_height()+0.3))
```
